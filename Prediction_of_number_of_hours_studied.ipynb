```
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sr

# Load the dataset
df = pd.read_csv('StudentPerformance.csv')
print(df.head())

# Preprocessing: Convert 'Extracurricular Activities' to numeric (1=Yes, 0=No)
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({'Yes': 1, 'No': 0})

# Define Features (X) and Target (y)
X = df.drop('Hours Studied', axis=1)
y = df['Hours Studied']

# --- Part 1: Hypothesis Testing (Statsmodels) ---
# Add constant for the intercept term
X_const = sm.add_constant(X)
model_sm = sm.OLS(y, X_const).fit()
print(model_sm.summary())

# --- Part 2: Prediction (Sklearn) ---
# Fit the sklearn model
model_sk = LinearRegression()
model_sk.fit(X, y)

# Create Unknown Data
unknown_data = pd.DataFrame({
    'Previous Scores': [85, 60, 90],
    'Extracurricular Activities': [1, 0, 1],
    'Sleep Hours': [7, 6, 8],
    'Sample Question Papers Practiced': [3, 1, 5],
    'Performance Index': [80, 50, 95]
})

# Make Predictions for unknown data
predictions_unknown = model_sk.predict(unknown_data)
print("Predictions for unknown data:", predictions_unknown)

# Calculate R-squared score for the model on the full dataset
r2_score_ = sr.r2_score(y, model_sk.predict(X))
print("R-squared score on full dataset:", r2_score_)
```
